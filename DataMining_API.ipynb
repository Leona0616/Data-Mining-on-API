```python
import time
import datetime
from datetime import timedelta, date
import requests
import json
import pandas
import csv
import urllib.parse
import pandas as pd
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry


#the SN number is the input
SN = ["AL2002017110051","AL2002017110097" ] #put your inputs here
limittime = datetime.datetime.strptime("2018-12-13", "%Y-%m-%d")
print('Start....')

for sn in SN:
    starttime = datetime.datetime.strptime("2018-12-10", "%Y-%m-%d")
    endtime = datetime.datetime.strptime("2018-12-11", "%Y-%m-%d")
    main_api = "http://..CONFIDENTIAL LINK." #the api you want to access
    changed_api = "sn={}"
    append_api = "&starttime={}&endtime={}" #can be deleted
        
    while True:
        starttime = starttime + datetime.timedelta(days=1)
        endtime = endtime + datetime.timedelta(days=1)
        #url = URL_pattern.format(starttime,endtime)
        url_loo = main_api+changed_api.format(sn)+append_api.format(starttime.strftime("%Y-%m-%d,%H:%M:%S"),endtime.strftime("%Y-%m-%d,%H:%M:%S"))
        print(url_loo)   
        response = requests.get(url_loo)
        print(response.status_code)
        
        if response.status_code == 200:
            trial_data = requests.get(url_loo).json()   
            #print the errors when request the url
            try:
                r = requests.get(url_loo,timeout=10)
                r.raise_for_status()
            except requests.exceptions.HTTPError as errh:
                print ("Http Error:",errh)
                continue
            except requests.exceptions.ConnectionError as errc:
                print ("Error Connecting:",errc)
                continue
            except requests.exceptions.Timeout as errt:
                print ("Timeout Error:",errt)
            except requests.exceptions.RequestException as err:
                print ("OOps: Something Else",err)
            except Exception:
                #sleep for 20 seconds in case that helps
                time.sleep(20)
                #try again
                get(url_loo)
            #save the dataset to .json file and make the data looks nice
            with open('trial2_data.json', 'w', encoding = 'utf-8') as f:
                json.dump(trial_data,f, ensure_ascii=False, indent = 4)
            trial_df = pd.read_json('trial2_data.json')

            #algorithm for each value        
            trial_df['PVgen'] = #ALGORITHM 
            trial_df['LOAD'] = #ALGORITHM
            trial_df['FN'] =#ALGORITHM

            trial_df.loc[trial_df['FN']>0,"FEEDIN"]= 0
            trial_df.loc[trial_df['FN']<=0,"FEEDIN"]= trial_df['FN'].abs()
            trial_df.loc[trial_df['LOAD']<0,"LOAD"]= 0
            #print(trial_df['FN'])

            trial_df.loc[trial_df['FN'] < 100,"GridCon"]= 0
            trial_df.loc[trial_df['FN'] >= 100,"GridCon"]= trial_df['FN']
            
            #Another way to calculate ternary condition by Numpy
            #trial_df['FEEDIN'] = np.where(trial_df.FN>0, 0, trial_df['FN'].abs())
            #trial_df['GridCon'] = np.where(trial_df.FN<100, 0, trial_df.FN)

            df = trial_df[['Sn','UploadTime','PVgen','LOAD','FEEDIN','GridCon','Soc']]

            #save the data to CSV
            with open('trial_output_v15.csv','a') as fd:
                df.to_csv(fd, header=False)

            #Once the the endtime is greater than limittime, the loop will break
            if endtime >= limittime:
                print('NEXT SN')
                break
            continue
        else:
            trial_data = requests.get(url_loo).json()
            print('Wrong Connection')
        
else:
    print('Loop Done')

```
